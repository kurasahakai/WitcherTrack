```python
import numpy as np
import cv2
import glob
from PIL import Image
from IPython.display import display

images = glob.glob('tests/fixtures/*.png') + glob.glob('tests/fixtures/*.jpg')
```

```python
def preprocess(path):    
    pic = cv2.imread(path)
    # pic = cv2.filter2D(pic, -1, np.array([[-1,-1,-1], [-1, 9,-1], [-1,-1,-1]]))
    pic = cv2.cvtColor(pic, cv2.COLOR_RGB2BGR)
    hsv = cv2.cvtColor(pic, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, (100, 50, 220), (150, 120, 255))
    
    pic = cv2.bitwise_and(pic, pic, mask=mask)
    pic = cv2.cvtColor(pic, cv2.COLOR_BGR2GRAY)
    pic = cv2.dilate(pic, None)
    # pic = cv2.filter2D(pic, -1, np.array([[-1,-1,-1], [-1, 9,-1], [-1,-1,-1]]))
    # _, pic = cv2.threshold(pic, 140, 255, cv2.THRESH_BINARY)
   
    #, pic = cv2.threshold(pic, 140, 255, cv2.THRESH_BINARY)
    
    # 30 30 90
    # 35 40 100
    display(Image.fromarray(pic))
    
for i in images:
    preprocess(i)
```
